## TRANSFORMING THE VARIABLES

```python
df1_clean['horsepower_sq'] = df1_clean['horsepower']**2
df1_clean['price_log'] = np.log(df1_clean['price']) 
```

```python
plt.scatter(df1_clean['horsepower_sq'] , df1_clean['price_log'])
```

```python
reg2 = LinearRegression()
reg2.fit(df1_clean['horsepower_sq'].values.reshape(-1, 1) , df1_clean['price_log'].values.reshape(-1, 1))
```

```python
results2 = reg1.score(df1_clean['horsepower_sq'].values.reshape(-1, 1) , df1_clean['price_log'].values.reshape(-1, 1))
results2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming reg2 is your LinearRegression model
# Assuming df1_clean is your DataFrame

# Make predictions using the model
df1_clean['price_log_pred'] = reg2.predict(df1_clean['horsepower_sq'].values.reshape(-1, 1))

# Plot the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower_sq', y='price_log', data=df1_clean, label='Original Data')

# Plot the regression line
sns.lineplot(x='horsepower_sq', y='price_log_pred', data=df1_clean, color='red', label='Regression Line')

plt.xlabel('Horsepower Squared')
plt.ylabel('Log-Transformed Price')
plt.title('Scatter Plot with Regression Line')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

y_pred = reg2.predict(df1_clean['horsepower_sq'].values.reshape(-1, 1))

# Calculate R-squared value
r2_value = r2_score(df1_clean['price_log'], y_pred)

print(f'R-squared value: {r2_value}')
```

## ANOVA TEST FOR OVERALL MODEL IMPORTANCE

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from scipy.stats import f

# Assuming mlr_st is your standardized feature matrix and y is your target variable
reg3 = LinearRegression()
model3 = reg3.fit(mlr_st, y)

# Number of observations
n = len(y)

# Number of predictors
p = mlr_st.shape[1]

# Degrees of freedom for the numerator (df_regression)
df_regression = p - 1

# Degrees of freedom for the denominator (df_residual)
df_residual = n - p

# Calculate the F-statistic
f_statistic = (np.sum((model3.predict(mlr_st) - np.mean(y))**2) / df_regression) / (np.sum((y - model3.predict(mlr_st))**2) / df_residual)

# Calculate the p-value
p_value = 1 - f.cdf(f_statistic, df_regression, df_residual)

print("F-statistic:", f_statistic)
print("p-value:", p_value)
```
